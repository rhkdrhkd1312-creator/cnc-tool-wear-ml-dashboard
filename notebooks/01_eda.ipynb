```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path('..').resolve()))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.load_data import load_train, load_all_experiments
from src.preprocess import apply_doc_rule

train = load_train()
experiments = load_all_experiments()
train.head()
```

```python
pd.crosstab(train.tool_condition, train.feed_rate)
sns.histplot(train, x='clamp_pressure', hue='tool_condition', multiple='dodge')
plt.show()
```

```python
rates = []
for eid, df in experiments.items():
    rates.append({'experiment_id': eid, 'drop_rate': 1-len(apply_doc_rule(df))/len(df)})
pd.DataFrame(rates).merge(train[['experiment_id','tool_condition']], on='experiment_id').head(10)
```
